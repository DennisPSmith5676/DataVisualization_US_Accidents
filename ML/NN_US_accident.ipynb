```python
# Create a LabelEncode instance
le = LabelEncoder()

# Fit and transform the LabelEncode using the categorical variable list
accident_df.Weekday  = le.fit_transform(accident_df.Weekday)
accident_df.Weather_Condition  = le.fit_transform(accident_df.Weather_Condition)
accident_df.Sunrise_Sunset  = le.fit_transform(accident_df.Sunrise_Sunset)
accident_df.City  = le.fit_transform(accident_df.City)
accident_df.County  = le.fit_transform(accident_df.County)
accident_df.State  = le.fit_transform(accident_df.State)
accident_df.Timezone  = le.fit_transform(accident_df.Timezone)
accident_df.Wind_Direction  = le.fit_transform(accident_df.Wind_Direction)
accident_df.Month  = le.fit_transform(accident_df.Month)

accident_df
```

```python
# Split our preprocessed data into our features and target arrays
y=accident_df["Severity"].values
X=accident_df.drop(["Severity"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = number_input_features*2
hidden_nodes_layer2 = number_input_features

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="relu"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Import checkpoint Dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

# Define checkpoint path and filenmaes
os.makedirs("checkpoints/",exist_ok= True)
checkpoint_path="checkpoints/weights.{epoch:02d}.hdf5"
```

```python
# Create a callback that saves the model's weights every epoch
cp_callback=ModelCheckpoint(
filepath=checkpoint_path,
verbose=1,
save_weights_only=True,
save_freq='epoch',
period=5
)
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=25, callbacks=[cp_callback])
```
